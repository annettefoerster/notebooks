## 3D Rankine Vortex

```julia
radii = [1:1:300]
vmax = 40
rmw = 40
vt = [[vmax*i/rmw for i in radii[1:rmw-1]],[vmax*(rmw/i)^(0.4) for i in radii[rmw:end]]]
println(vt[38:1:42]')
vts = Dict()
vts[1] = vt
for k in [2:1:10] 
  vts[k] = vt .*(cos(k/180.0*pi*8))
end

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
 plt.plot(radii,vts[k],label=string(k)*" km")
end
plt.legend(fontsize=6)
plt.xlabel("radius (km)")
plt.ylabel("Tangential wind (m/s)")
```

## Calculating p(r,z)

```julia
function calc_dz(dict::Dict,dz)
  dvtdz = Dict()
  kend = length(dict)    
  for k in [2:1:kend-1]
      dvtdz[k] = (dict[k+1] .- dict[k-1]) ./(2*dz)
  end
  dvtdz[1] = (dict[2] .- dict[1]) ./dz
  dvtdz[kend] = (dict[kend] .- dict[kend-1]) ./dz
  dvtdz    
end

dz = 1000.0
dvtdz = calc_dz(vts,dz)

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
    plt.plot(radii,dvtdz[k],label=string(k)*" km")
end
plt.legend(fontsize=6)
plt.xlabel("radius (km)")
plt.ylabel("Dvtdz (m/s)")
```

```julia
# Vertical profiles for a adiabatic atmosphere
H = 29.5; cv = 718; R = 287.0; rho_zero = 1.2; p_zero = 1013
rho_z = Dict()
p_z = Dict()
for k in [1:1:10]
  rho_z[k] = rho_zero*(1-k/H)^(cv/R)
  p_z[k] = p_zero*(1-k/H)^((cv+R)/R)
end
fig = plt.figure(figsize=(3, 3))
rho_plot = [rho_z[k] for k in [1:1:10]]
plt.plot(rho_plot,[1:1:10])
plt.xlabel("density (kg/m^3)")
plt.ylabel("altitude (km)")
```

```julia
f = 0.0000585

function calc_dpdr(radii,vt,rho)
    dpdr = Array(Number,(length(radii)))
    dp = Array(Number,(length(radii)))
    for i in [1:1:length(radii)]
        r = radii[i]*1000.0
        v = vt[i]
        dpdr[i] = rho*(v^2/r +f*v)     
    end
    dpdr 
end

dpdr = Dict()
dp = Dict()
for k in [1:1:10]
    dpdr[k] = calc_dpdr(radii,vts[k],rho_z[k])
    dp[k] = dpdr[k] .*1000.0 ./100.0
end

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
    plt.plot(radii,dpdr[k],label=k)
end
plt.legend(fontsize=6)
plt.xlabel("radius (km)")
plt.ylabel("dpdr (hPa)")
```

```julia
function sum_dp(radii,dp)
    delta_p = Array(Number,(length(radii)))
    dp_sum = 0
    for i in reverse([1:1:length(radii)])
        dp_sum += dp[i]
        delta_p[i] = dp_sum
    end
    delta_p
end

delta_p = Dict()
for k in [1:1:10]
    delta_p[k] = sum_dp(radii,dp[k])
end

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
    plt.plot(radii,-delta_p[k],label=k)
end
plt.legend(fontsize=6)
plt.xlabel("radius (km)")
plt.ylabel("integrated pressure deficit (hPa)")
```

```julia
p = Dict()
for k in [1:1:10]
    p[k] = delta_p[k] .*(-1) .+p_z[k]
end

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
    plt.plot(radii,p[k],label=k)
end
plt.legend(fontsize=6)
plt.xlabel("radius (km)")
plt.ylabel("pressure (hPa)")
```

## Calculate dρdr

```julia
#drhodr = -1/g d/dz (rho_0 (v^2/r +f*v))
g = 9.81
dz = 1000.0

rhoCs = Dict()
for k in [1:1:10]
    vt = vts[k]
    rho = rho_z[k]
    rhoC = [rho*((v^2/r +f*v)) for (r,v) in zip(radii .*1000.0,vt)]  
    rhoCs[k] = rhoC
end
ddz = calc_dz(rhoCs,dz)

drhodr = Dict()
drho = Dict()
for k in [1:1:10]
    drhodr[k] = ddz[k] .*(-1/g)
    drho[k] = drhodr[k].*1000.0
end

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
    plt.plot(radii,drhodr[k],label=k)
end
plt.legend(fontsize=6)
plt.xlabel("radius (km)")
plt.ylabel("drhodr (kg/m^3)")
```

```julia
function sum_drho(radii,drho)
    delta_rho = Array(Number,(length(radii)))
    drho_sum = 0
    for i in reverse([1:1:length(radii)])
        drho_sum += drho[i]
        delta_rho[i] = drho_sum
    end
    delta_rho
end

delta_rho = Dict()
for k in [1:1:10]
    delta_rho[k] = sum_drho(radii,drho[k])
end

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
    plt.plot(radii,-delta_rho[k],label=k)
end
plt.legend(fontsize=6)
sum(delta_rho[1])
plt.xlabel("radius (km)")
plt.ylabel("integrated density deficit (kg/m^3)")
```

```julia
rho = Dict()
for k in [1:1:10]
    rho[k] = -delta_rho[k] .+rho_z[k]
end

fig = plt.figure(figsize=(3, 3))
for k in [1:1:10]
    plt.plot(radii,rho[k],label=k)
end
plt.legend(fontsize=6)
plt.xlabel("radius (km)")
plt.ylabel("density (kg/m^3)")
```
